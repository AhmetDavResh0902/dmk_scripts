```python
import pandas as pd
import csv
from ast import literal_eval
import matplotlib.pyplot as plt
import numpy as np
from os import path
import pathlib
```

```python
df2017 = pd.read_csv(r"C:\Users\ahmet\YandexDisk\ДИПЛОМ\Материалы\year_2017.csv")
df2018 = pd.read_csv(r"C:\Users\ahmet\YandexDisk\ДИПЛОМ\Материалы\year_2018.csv")
df2019 = pd.read_csv(r"C:\Users\ahmet\YandexDisk\ДИПЛОМ\Материалы\year_2019.csv")
df2020 = pd.read_csv(r"C:\Users\ahmet\YandexDisk\ДИПЛОМ\Материалы\year_2020.csv")
df2021 = pd.read_csv(r"C:\Users\ahmet\YandexDisk\ДИПЛОМ\Материалы\year_2021.csv")
```

```python
dfs2017 = df2017[["name", "key_skills", "description"]]
dfs2017['name'] = dfs2017['name'].astype(str).str.lower()
dfs2018 = df2018[["name", "key_skills", "description"]]
dfs2018['name'] = dfs2018['name'].astype(str).str.lower()
dfs2019 = df2019[["name", "key_skills", "description"]]
dfs2019['name'] = dfs2019['name'].astype(str).str.lower()
dfs2020 = df2020[["name", "key_skills", "description"]]
dfs2020['name'] = dfs2020['name'].astype(str).str.lower()
dfs2021 = df2021[["name", "key_skills", "description"]]
dfs2021['name'] = dfs2021['name'].astype(str).str.lower()
```

```python
dfs_all = pd.concat([dfs2017, dfs2018, dfs2019, dfs2020, dfs2021])
dfs_all
```

### ПОДКЛЮЧЕНИЕ ФАЙЛА СО СКИЛЛАМИ

```python
soft = literal_eval(open(r"C:\Users\ahmet\YandexDisk\Словари скиллов\soft.txt", "r", encoding='utf-8').read())
hard = literal_eval(open(r"C:\Users\ahmet\YandexDisk\Словари скиллов\hard.txt", "r", encoding='utf-8').read())
old_hard = literal_eval(open(r"C:\Users\ahmet\YandexDisk\Словари скиллов\group_old.txt", "r", encoding='utf-8').read())
new_hard = literal_eval(open(r"C:\Users\ahmet\YandexDisk\Словари скиллов\group03052022.txt", "r", encoding='utf-8').read())
```

```python
import os
path = r"C:\Users\ahmet\Documents\Anaconda\Не программисты" 
files = os.listdir(path)
txt = [i for i in files if i.endswith('.txt')]
txt
```

```python
import os
for x in txt:
    p = os.path.abspath(x)
    vac_str = literal_eval(open(p, "r", encoding='utf-8').read())
    s = os.path.basename(format(p))
    s = s.split(".txt")
    filename = s[0]

    # Работа с софт-скиллами
    df1 = dfs_all[dfs_all['name'].isin(vac_str)]
    df1 = df1['key_skills'].value_counts().to_frame()
    df1.reset_index(level=0, inplace = True)
    df1.rename(columns = {'index' : 'skills', 'key_skills' : 'nums'}, inplace = True)
    df1 = df1.skills.str.split("\n",expand=True).set_index(df1.nums).stack().reset_index(level=1, drop=True).reset_index().rename(columns={0:"skills"})
    df1 = df1.reindex(columns=['skills', 'nums'])
    df1 = df1['skills'].value_counts().to_frame()
    df1.reset_index(level=0, inplace = True)
    df1.rename(columns = {'index' : 'skills', 'skills' : 'nums'}, inplace = True)
    #Группировка скиллов для 2017 года
    sum_group = []
    group = list(soft.keys()) 
    
    for key in soft.keys():
        sum_gr = df1[df1['skills'].isin(soft[key])]
        sum_gr  = sum_gr.nums.sum()
        sum_group.append(sum_gr)
    df1 = pd.DataFrame(list(zip(group, sum_group)), columns =['Группа скиллов', 'Частотность']).sort_values(by='Частотность', ascending=False)
    df1 = df1.loc[df1['Частотность'] != 0]
    
    # Работа с хард-скиллами
    df2 = dfs_all[dfs_all['name'].isin(vac_str)]
    df2 = df2['key_skills'].value_counts().to_frame()
    df2.reset_index(level=0, inplace = True)
    df2.rename(columns = {'index' : 'skills', 'key_skills' : 'nums'}, inplace = True)
    df2 = df2.skills.str.split("\n",expand=True).set_index(df2.nums).stack().reset_index(level=1, drop=True).reset_index().rename(columns={0:"skills"})
    df2 = df2.reindex(columns=['skills', 'nums'])
    df2 = df2['skills'].value_counts().to_frame()
    df2.reset_index(level=0, inplace = True)
    df2.rename(columns = {'index' : 'skills', 'skills' : 'nums'}, inplace = True)
    #Группировка скиллов для 2018 года
    sum_group = []
    group = list(hard.keys()) 
    
    for key in hard.keys():
        sum_gr = df2[df2['skills'].isin(hard[key])]
        sum_gr  = sum_gr.nums.sum()
        sum_group.append(sum_gr)
    df2 = pd.DataFrame(list(zip(group, sum_group)), columns =['Группа скиллов', 'Частотность']).sort_values(by='Частотность', ascending=False)
    df2 = df2.loc[df2['Частотность'] != 0]
    
    # Работа со старыми хард-скиллами
    df3 = dfs_all[dfs_all['name'].isin(vac_str)]
    df3 = df3['key_skills'].value_counts().to_frame()
    df3.reset_index(level=0, inplace = True)
    df3.rename(columns = {'index' : 'skills', 'key_skills' : 'nums'}, inplace = True)
    df3 = df3.skills.str.split("\n",expand=True).set_index(df3.nums).stack().reset_index(level=1, drop=True).reset_index().rename(columns={0:"skills"})
    df3 = df3.reindex(columns=['skills', 'nums'])
    df3 = df3['skills'].value_counts().to_frame()
    df3.reset_index(level=0, inplace = True)
    df3.rename(columns = {'index' : 'skills', 'skills' : 'nums'}, inplace = True)
    #Группировка скиллов для 2019 года
    sum_group = []
    group = list(old_hard.keys()) 
    
    for key in old_hard.keys():
        sum_gr = df3[df3['skills'].isin(old_hard[key])]
        sum_gr  = sum_gr.nums.sum()
        sum_group.append(sum_gr)
    df3 = pd.DataFrame(list(zip(group, sum_group)), columns =['Группа скиллов', 'Частотность']).sort_values(by='Частотность', ascending=False)
    df3 = df3.loc[df3['Частотность'] != 0]
    
    # Работа с новыми хард-скиллами(03.05.2022)
    df4 = dfs_all[dfs_all['name'].isin(vac_str)]
    df4 = df4['key_skills'].value_counts().to_frame()
    df4.reset_index(level=0, inplace = True)
    df4.rename(columns = {'index' : 'skills', 'key_skills' : 'nums'}, inplace = True)
    df4 = df4.skills.str.split("\n",expand=True).set_index(df4.nums).stack().reset_index(level=1, drop=True).reset_index().rename(columns={0:"skills"})
    df4 = df4.reindex(columns=['skills', 'nums'])
    df4 = df4['skills'].value_counts().to_frame()
    df4.reset_index(level=0, inplace = True)
    df4.rename(columns = {'index' : 'skills', 'skills' : 'nums'}, inplace = True)
    #Группировка скиллов для 2020 года
    sum_group = []
    group = list(new_hard.keys()) 
    
    for key in new_hard.keys():
        sum_gr = df4[df4['skills'].isin(new_hard[key])]
        sum_gr  = sum_gr.nums.sum()
        sum_group.append(sum_gr)
    df4 = pd.DataFrame(list(zip(group, sum_group)), columns =['Группа скиллов', 'Частотность']).sort_values(by='Частотность', ascending=False)
    df4 = df4.loc[df4['Частотность'] != 0]
    
    # Формирование документа:
    salary_sheets = {'soft-скиллы': df1, 'hard-скиллы': df2, 'Старые hard': df3, 'Новые hard': df4}
    writer = pd.ExcelWriter(r'C:\Users\ahmet\Documents\Anaconda\Не программисты\Общая.' + filename + '.xlsx', engine='xlsxwriter')
    for sheet_name in salary_sheets.keys():
        salary_sheets[sheet_name].to_excel(writer, sheet_name=sheet_name, index=False)
    writer.save()
```
